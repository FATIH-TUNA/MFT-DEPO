```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
veri=pd.read_excel("kirmizi_sarap.xlsx")
veri.head()
```

```python
veri.info()
```

```python
veri.describe().T
```

```python
x=veri.drop("quality",axis=1)
y=veri["quality"]
x_egitim,x_test,y_egitim,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
scaler=StandardScaler()
x_egitim=scaler.fit_transform(x_egitim)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
model_mlp=MLPClassifier()
param_grid={'hidden_layer_sizes':(100,),'max_iter':[50,100,150],'activation':['tanh','relu'],'solver':['adam'],'alpha':[0.0001,0.05],'learning_rate':['constant','adapitve']}
scoring={'Accuracy':'accuracy'}
mlp=GridSearchCV(model_mlp,return_train_score=True,param_grid=param_grid,scoring=scoring,cv=5,refit='Accuracy')
mlp.fit(x_egitim,y_egitim)
```

```python
print("\nOPTIMUM SKOR: %.2f"%mlp.best_score_)
```

```python
from sklearn.metrics import*
tahmin_veri=mlp.best_estimator_.predict(x_egitim)
print("\nEGITIM MLP ICIN DOGRULUK ORANI: %.2f"%accuracy_score(y_egitim,tahmin_veri))
```

```python
tahmin_veri_test=mlp.best_estimator_.predict(x_test)
print("\nTEST MLP DOGRULUK ORANI: %.2f"%accuracy_score(y_test,tahmin_veri_test))
```

```python
print(classification_report(y_test,tahmin_veri_test))
```
