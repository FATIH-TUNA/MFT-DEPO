```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
from sklearn.model_selection import train_test_split,KFold,cross_val_score
```

```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
veri=pd.read_csv(url,sep=',',names=['age','workclass','fnlwgt','education','education-num','martial-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','ccountry','target'])
veri.head()
```

```python
veri.info()
```

```python
veri.describe().T
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for k in veri.columns:
    if veri[k].dtypes=='object':
        veri[k]=le.fit_transform(veri[k])
```

```python
veri.head()
```

```python
x=veri.drop('target',axis=1)
y=veri['target']
x_egitim,x_test,y_egitim,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import*
from xgboost import XGBRegressor
```

```python
model=[]
model.append(('model_nb',GaussianNB()))
model.append(('model_knn',KNeighborsClassifier()))
model.append(('model_lr',LogisticRegression()))
model.append(('model_mlp',MLPClassifier()))
```

```python
for name,i in model:
    kfold=KFold(n_splits=10,random_state=0,shuffle=True)
    sonuclar=cross_val_score(i,x_egitim,y_egitim,cv=kfold,scoring='accuracy')
    print(name," ICIN KFOLD SONUCLARI: ",sonuclar.mean())
    
```

```python
for name,k in model:
    k.fit(x_egitim,y_egitim)
    tahmin_test=k.predict(x_test)
    print(name," ICIN TAHMIN TEST SKORLARI: %.2f"%accuracy_score(y_test,tahmin_test))
```
