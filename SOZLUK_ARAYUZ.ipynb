```python
from PyQt5.QtCore import*
from PyQt5.QtGui import*
from PyQt5.QtWidgets import*
import sys
```

```python
class deneme(QDialog):
    def __init__(self):
        super().__init__()
        self.izgara=QGridLayout()
        self.sozluk={'elma':'apple','portakal':'orange','muz':'banana','inek':'cow','sut':'milk'}
        self.buton1=QPushButton("ARAMA")
        self.buton1.clicked.connect(self.tikla1)
        self.izgara.addWidget(self.buton1,0,0)
        
        self.buton2=QPushButton("EKLEME")
        self.buton2.clicked.connect(self.tikla2)
        self.izgara.addWidget(self.buton2,1,0)
        
        self.buton3=QPushButton("SILME")
        self.buton3.clicked.connect(self.tikla3)
        self.izgara.addWidget(self.buton3,2,0)
        
        self.buton4=QPushButton("LISTELE")
        self.buton4.clicked.connect(self.tikla4)
        self.izgara.addWidget(self.buton4,3,0)
        
        self.izgara.addWidget(QLabel("SONUC: "),15,0)
        self.sonuc=QLabel("<font color='red'>KELIME...</font>")
        self.izgara.addWidget(self.sonuc,15,1)
        
        self.setLayout(self.izgara)
        self.setWindowTitle("MFT")
        self.setGeometry(100,100,400,400)
    def tikla1(self):
        self.izgara.addWidget(QLabel("ARANACAK KELIME: "),4,0)
        self.kelime=QLineEdit()
        self.izgara.addWidget(self.kelime,4,1)
        
        self.buton5=QPushButton("ARA")
        self.buton5.clicked.connect(self.tikla5)
        self.izgara.addWidget(self.buton5,5,0)
    def tikla2(self):
        self.izgara.addWidget(QLabel("TURKCE KELIME: "),6,0)
        self.t_kelime=QLineEdit()
        self.izgara.addWidget(self.t_kelime,6,1)
        
        self.izgara.addWidget(QLabel("KELIMENIN INGILIZCESI: "),7,0)
        self.i_kelime=QLineEdit()
        self.izgara.addWidget(self.i_kelime,7,1)
        
        self.buton9=QPushButton("EKLE")
        self.buton9.clicked.connect(self.tikla9)
        self.izgara.addWidget(self.buton9,8,0)
    def tikla9(self):
        self.kelime_t=str(self.t_kelime.text())
        self.kelime_i=str(self.i_kelime.text())
        self.sozluk[self.kelime_t]=self.kelime_i
        self.sonuc.setText(str(self.sozluk))
    
    def tikla3(self):
        self.sonuc.setText(str(self.sozluk))
        self.izgara.addWidget(QLabel("SILINECEK KELIME: "),9,0)
        self.silinecek=QLineEdit()
        self.izgara.addWidget(self.silinecek,9,1)
        
        self.buton10=QPushButton("SILINECEK")
        self.buton10.clicked.connect(self.tikla10)
        self.izgara.addWidget(self.buton10,10,0)
    def tikla10(self):
        self.sil=str(self.silinecek.text())
        if self.sil in self.sozluk:
            del self.sozluk[self.sil]
            self.sonuc.setText(str(self.sozluk))
        else:
            self.mesaj=QMessageBox()
            self.mesaj.setIcon(QMessageBox.Warning)
            self.mesaj.setWindowTitle("UYARI MESAJI...")
            self.mesaj.setText("BU KELIME SOZLUKTE BULUNAMADI...")
            self.mesaj.setStandardButtons(QMessageBox.Ok)
            self.izgara.addWidget(self.mesaj,12,0)
    
    def tikla4(self):
        self.sonuc.setText(str(self.sozluk))
    
    def tikla5(self):
        self.sozcuk=str(self.kelime.text())
        if self.sozcuk in self.sozluk:
            self.sonuc.setText(str(self.sozluk[self.sozcuk]))
        elif self.sozcuk not in self.sozluk:
            self.sonuc.setText("KELIMENIN ANLAMI BUUNAMADI...")
            self.buton6=QPushButton("EKLE")
            self.buton6.clicked.connect(self.tikla6)
            self.izgara.addWidget(self.buton6,8,0)
            
            self.buton7=QPushButton("VAZGEC")
            self.buton7.clicked.connect(self.tikla7)
            self.izgara.addWidget(self.buton7,9,0)
            
    def tikla6(self):
        self.izgara.addWidget(QLabel("KELIMEIN INGILIZCESI: "),10,0)
        self.ingilizce=QLineEdit()
        self.izgara.addWidget(self.ingilizce,10,1)
        
        self.buton8=QPushButton("ONAYLA")
        self.buton8.clicked.connect(self.tikla8)
        self.izgara.addWidget(self.buton8,11,0)
    def tikla8(self):
        self.ingi=str(self.ingilizce.text())
        self.sozluk[self.sozcuk]=self.ingi
        self.sonuc.setText(str(self.sozluk))
        
        
    def tikla7(self):
        self.sonuc.setText(str(self.sozluk))
        self.mesaj=QMessageBox()
        self.mesaj.setIcon(QMessageBox.Warning)
        self.mesaj.setWindowTitle("UYARI MESAJI...")
        self.mesaj.setText("BULUNAMAYAN KELIME SOZLUGE EKLENMEDI...")
        self.mesaj.setStandardButtons(QMessageBox.Ok)
        self.izgara.addWidget(self.mesaj,12,0)

app=QApplication(sys.argv)
pencere=deneme()
pencere.show()
sys.exit(app.exec_())
```
