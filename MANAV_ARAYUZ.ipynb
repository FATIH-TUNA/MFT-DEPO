```python
from PyQt5.QtGui import*
from PyQt5.QtCore import*
from PyQt5.QtWidgets import*
import sys


class deneme(QDialog):
    def __init__(self):
        super().__init__()
        self.liste=[]
        self.liste_meyve={'meyveler':{'ELMA':4,'MUZ':12,'NAR':25}}
        self.liste_sebze={'sebzeler':{'DOMATES':4,'PATLICAN':2,'BIBER':5}}
        self.izgara=QGridLayout()
        
        self.buton1=QPushButton("SATICI")
        self.izgara.addWidget(self.buton1,0,0)
        self.buton1.clicked.connect(self.tikla1)
        
        self.buton2=QPushButton("MUSTERI")
        self.izgara.addWidget(self.buton2,1,0)
        self.buton2.clicked.connect(self.tikla2)
        
        self.izgara.addWidget(QLabel("URUNLER: "),16,0)
        self.sonuc=QLabel("<font color='red'>urun bolumu...</font>")
        self.izgara.addWidget(self.sonuc,16,1)
        
        self.izgara.addWidget(QLabel("FIYAT: "),17,0)
        self.sonuc2=QLabel("<font color='red'>URUN FIYATI....<font>")
        self.izgara.addWidget(self.sonuc2,17,1)
        
        
        self.setLayout(self.izgara)
        self.setWindowTitle("MFT")
        self.setGeometry(100,100,400,400)
    def tikla1(self):
        
        self.izgara.addWidget(QLabel("KULLANICI: "),2,0)
        self.kullanici=QLineEdit()
        self.izgara.addWidget(self.kullanici,2,1)
        
        self.izgara.addWidget(QLabel("PAROLA: "),3,0)
        self.sifre=QLineEdit()
        self.izgara.addWidget(self.sifre,3,1)
        
        self.buton3=QPushButton("GIRIS")
        self.izgara.addWidget(self.buton3,4,0)
        self.buton3.clicked.connect(self.tikla3)
        
    def tikla3(self):
        self.ad=str(self.kullanici.text())
        self.parola=int(self.sifre.text())
        if(self.ad=='MEHMET' and self.parola==12345):
            self.buton4=QPushButton("MEYVELER")
            self.izgara.addWidget(self.buton4,5,0)
            self.buton4.clicked.connect(self.tikla4)
            
            self.buton5=QPushButton("SEBZELER")
            self.izgara.addWidget(self.buton5,6,0)
            self.buton5.clicked.connect(self.tikla5)
            
            
        elif(self.ad!='MEHMET' and self.parola==12345):
            self.mesaj=QMessageBox()
            self.mesaj.setIcon(QMessageBox.Warning)
            self.mesaj.setWindowTitle("KULLANICI ADI HATALI GIRILDI...")
            self.mesaj.setText("LUTFEN KULLANICI ADINI YENIDEN GIRINIZ...")
            self.mesaj.setStandardButtons(QMessageBox.Ok)
            self.izgara.addWidget(self.mesaj,17,0)
            
            
        elif(self.ad=='MEHMET' and self.parola!=12345):
            self.mesaj=QMessageBox()
            self.mesaj.setIcon(QMessageBox.Warning)
            self.mesaj.setWindowTitle("PAROLA HATALI GIRILDI...")
            self.mesaj.setText("LUTFEN PPAROLAYI TEKRAR GIRINIZ...")
            self.mesaj.setStandardButtons(QMessageBox.Ok)
            self.izgara.addWidget(self.mesaj,17,0)
            
            
            
        elif(self.ad!='MEHMET' and self.parola!=12345):
            self.mesaj=QMessageBox()
            self.mesaj.setIcon(QMessageBox.Warning)
            self.mesaj.setWindowTitle("KULLANICI ADI VE PAROLA HATALI...")
            self.mesaj.setText("KULANICI ADI VE PAROLAYI TEKRAR GIRINIZ...")
            self.izgara.addWidget(self.mesaj,17,0)
            
    def tikla4(self):
        
        self.buton6=QPushButton("EKLE")
        self.izgara.addWidget(self.buton6,7,0)
        self.buton6.clicked.connect(self.tikla6)
        
        self.buton7=QPushButton("GUNCELLE")
        self.izgara.addWidget(self.buton7,8,0)
        self.buton7.clicked.connect(self.tikla7)
    
    def tikla5(self):
        self.buton10=QPushButton("EKLE")
        self.izgara.addWidget(self.buton10,7,0)
        self.buton10.clicked.connect(self.tikla10)
        
        self.buton11=QPushButton("GUNCELLE")
        self.izgara.addWidget(self.buton11,8,0)
        self.buton11.clicked.connect(self.tikla11)
    def tikla10(self):
        self.izgara.addWidget(QLabel("SEBZE ADI: "),9,0)
        self.sebze=QLineEdit()
        self.izgara.addWidget(self.sebze,9,1)
        
        self.izgara.addWidget(QLabel("SEBZE FIYATI: "),10,0)
        self.sebze_fiyat=QLineEdit()
        self.izgara.addWidget(self.sebze_fiyat,10,1)
        
        self.buton12=QPushButton("EKLE")
        self.izgara.addWidget(self.buton12,11,0)
        self.buton12.clicked.connect(self.tikla12)
        
    def tikla11(self):
        self.izgara.addWidget(QLabel("SEBZE ADI: "),12,0)
        self.ad_sebze=QLineEdit()
        self.izgara.addWidget(self.ad_sebze,12,1)
        
        self.izgara.addWidget(QLabel("GUNCEL FIYAT: "),13,0)
        self.fiyat_guncelle=QLineEdit()
        self.izgara.addWidget(self.fiyat_guncelle,13,1)
        
        self.buton13=QPushButton("GUNCELLE")
        self.izgara.addWidget(self.buton13,14,0)
        self.buton13.clicked.connect(self.tikla13)
        
    def tikla12(self):
        self.sebze_adi=str(self.sebze.text())
        self.sebze_fiyati=float(self.sebze_fiyat.text())
        self.liste_sebze['sebzeler'][self.sebze_adi]=self.sebze_fiyati
        self.sonuc.setText(str(self.liste_sebze))
        
    def tikla13(self):
        self.ad_s=str(self.ad_sebze.text())
        self.s_fiyat=float(self.fiyat_guncelle.text())
        self.liste_sebze['sebzeler'][self.ad_s]=self.s_fiyat
        self.sonuc.setText(str(self.liste_sebze))
        
    def tikla6(self):
        
        self.izgara.addWidget(QLabel("URUN ADI: "),9,0)
        self.urun_1=QLineEdit()
        self.izgara.addWidget(self.urun_1,9,1)
        
        self.izgara.addWidget(QLabel("URUN FIYATI: "),10,0)
        self.urun_fiyat=QLineEdit()
        self.izgara.addWidget(self.urun_fiyat,10,1)
        
        self.buton8=QPushButton("ONAYLA")
        self.izgara.addWidget(self.buton8,11,0)
        self.buton8.clicked.connect(self.tikla8)
        
    def tikla7(self):
        self.izgara.addWidget(QLabel("URUN ADI: "),12,0)
        self.guncel_urun=QLineEdit()
        self.izgara.addWidget(self.guncel_urun,12,1)
        
        self.izgara.addWidget(QLabel("GUNCEL FIYAT: "),13,0)
        self.guncel_fiyat=QLineEdit()
        self.izgara.addWidget(self.guncel_fiyat,13,1)
        
        self.buton9=QPushButton("GUNCELLE")
        self.izgara.addWidget(self.buton9,14,0)
        self.buton9.clicked.connect(self.tikla9)
    
    def tikla8(self):
        self.urun=str(self.urun_1.text())
        self.fiyat=float(self.urun_fiyat.text())
        self.liste_meyve['meyveler'][self.urun]=self.fiyat
        self.sonuc.setText(str(self.liste_meyve))
    def tikla9(self):
        self.g_urun=str(self.guncel_urun.text())
        self.g_fiyat=float(self.guncel_fiyat.text())
        self.liste_meyve['meyveler'][self.g_urun]=self.g_fiyat
        self.sonuc.setText(str(self.liste_meyve))
        
    def tikla2(self):
        
        self.label=QLabel("MEYVELER")
        self.combo=QComboBox()
        self.combo.addItem("URUN SECINIZ...")
        self.combo.addItem("ELMA")
        self.combo.addItem("MUZ")
        self.combo.addItem("CILEK")
        self.combo.addItem("PORTAKAL")
        self.combo.addItem("NAR")
        self.izgara.addWidget(self.combo,5,1)
        self.izgara.addWidget(self.label,5,0)
        
        self.label1=QLabel("SEBZELER: ")
        self.izgara.addWidget(self.label1,6,0)
        self.combo1=QComboBox()
        self.combo1.addItem("URUN SECINIZ...")
        self.combo1.addItem("DOMATES")
        self.combo1.addItem("PATATES")
        self.izgara.addWidget(self.combo1,6,1)
        
        self.izgara.addWidget(QLabel("URUN ADI GIRINIZ: "),7,0)
        self.urunn=QLineEdit()
        self.izgara.addWidget(self.urunn,7,1)
    
        self.izgara.addWidget(QLabel("MIKTAR GIRINIZ: "),8,0)
        self.miktar=QLineEdit()
        self.izgara.addWidget(self.miktar,8,1)
        
        self.izgara.addWidget(QLabel("BIRIM FIYAT: "),9,0)
        self.birim=QLineEdit()
        self.izgara.addWidget(self.birim,9,1)
        
        self.buton=QPushButton("HESAPLA")
        self.buton.clicked.connect(self.tikla15)
        self.izgara.addWidget(self.buton,2,0)
    def tikla15(self):
        self.dene1=str(self.combo)
        self.dene2=str(self.combo1)
        self.urun=str(self.urunn.text())
        self.kilo=float(self.miktar.text())
        self.birimm=float(self.birim.text())
        self.hesap=self.kilo*self.birimm
        self.liste.append(self.hesap)
        self.toplam=sum(self.liste)
        self.sonuc.setText(str(self.hesap))
        self.sonuc2.setText(str(self.toplam))
        if self.urun not in self.dene1 or self.dene2:
            self.sonuc2.setText("URUN TUKENMISTIR...")
            
        
app=QApplication(sys.argv)
pencere=deneme()
pencere.show()
sys.exit(app.exec_())
```
